```python
import pandas as pd
from pathlib import Path
from Bio import Phylo
```

```python
# Name gene families using human gene symboles 
HOG = pd.read_csv('./HOG.tsv', sep = '\t', header=0)
HOG = HOG[['HOG', 'HomSap']]
# If a gene family contains multiple genes, then the default choice is to use the first human gene symbol as the name for that gene family.
HOG = HOG[~HOG.HomSap.isna()]
HOG['HomSap'] = HOG['HomSap'].apply(lambda x: x.split(',')[0].replace('|HomSap', ''))
HOG
```

```python
# draw the phylogenetic tree
tree = Phylo.parse('./CAFEV5/Gamma_asr.tre', 'nexus')
a = next(tree)
tree_dic = {i.name:i for i in Phylo.parse('./CAFEV5/Gamma_asr.tre', 'nexus')}

for i in a.get_terminals():
    i.name = i.name.split('_')[0].replace('*','')
for i in a.get_nonterminals():
    i.name = i.name.split('_')[0].replace('*', '')
Phylo.draw(a)
```

```python
family_results = pd.read_csv('./CAFEV5/Gamma_clade_results.txt', sep='\t', header=0)
family_results
```

```python
# load the pvalue of gene families
pvalue = pd.read_csv('./CAFEV5/Gamma_branch_probabilities.tab', header=0, sep='\t')

# load the change number of gene families
change = pd.read_csv('./CAFEV5/Gamma_change.tab', header=0, sep='\t')
```

```python
# Customize a function to filter out significantly changed gene families for each species. 
def sig_change(species, pvalue, change):
    sig = pvalue.loc[pvalue[species]<0.05, ['FamilyID', species]]
    exp = change.loc[change[species]>0, ['FamilyID', species]]
    con = change.loc[change[species]<0, ['FamilyID', species]]
    
    sig_exp = pd.merge(sig, exp, on='FamilyID', suffixes=['_pvalue', '_change'], how='inner')
    sig_con = pd.merge(sig, con, on='FamilyID', suffixes=['_pvalue', '_change'], how='inner')
    
    return sig_exp, sig_con
```

```python
tmp = list()
for i in change.columns[1:]:
    sig_exp, sig_con = sig_change(i, pvalue, change)
    tmp.append([i, len(sig_exp), len(sig_con)])
pd.DataFrame(tmp, columns=['SP', 'exp', 'con']).to_csv('test.tsv', header=True, index=False, sep='\t')
```

```python
# The evolved gene families of ancestor node of N. aviatro and P. kuhlii
sp_2_exp, sp_2_con = sig_change('<36>', pvalue=pvalue, change=change)
sp_2_exp = pd.merge(HOG, sp_2_exp, left_on='HOG', right_on='FamilyID', how='inner')
sp_2_con = pd.merge(HOG, sp_2_con, left_on='HOG', right_on='FamilyID', how='inner')
```

```python
# The significant evoloved gene families in N. aviator
NycAvi_sig = pvalue.loc[pvalue['NycAvi<17>']<0.05, ['FamilyID', 'NycAvi<17>']]
NycAvi_exp = change.loc[change['NycAvi<17>']>0, ['FamilyID', 'NycAvi<17>']]
NycAvi_con = change.loc[change['NycAvi<17>']<0, ['FamilyID', 'NycAvi<17>']]

NycAvi_sig_exp = pd.merge(NycAvi_sig, NycAvi_exp, on='FamilyID', suffixes=['_pvalue', '_change'], how='inner')
NycAvi_sig_con = pd.merge(NycAvi_sig, NycAvi_con, on='FamilyID', suffixes=['_pvalue', '_change'], how='inner')
```

```python
# The significant evoloved gene families in I. io
IaIo_sig = pvalue.loc[pvalue['IaIo<18>']<0.05, ['FamilyID', 'IaIo<18>']]
IaIo_exp = change.loc[change['IaIo<18>']>0, ['FamilyID', 'IaIo<18>']]
IaIo_con = change.loc[change['IaIo<18>']<0, ['FamilyID', 'IaIo<18>']]

IaIo_sig_exp = pd.merge(IaIo_sig, IaIo_exp, on='FamilyID', suffixes=['_pvalue', '_change'], how='inner')
IaIo_sig_con = pd.merge(IaIo_sig, IaIo_con, on='FamilyID', suffixes=['_pvalue', '_change'], how='inner')
```

```python
# using human symbol to annotate evolved gene families in N. aviator
NycAvi_sig_exp = pd.merge(HOG, NycAvi_sig_exp, left_on='HOG', right_on='FamilyID', how='inner')
NycAvi_sig_con = pd.merge(HOG, NycAvi_sig_con, left_on='HOG', right_on='FamilyID', how='inner')

# using human symbol to annotate evolved gene families in I. io
IaIo_sig_exp = pd.merge(HOG, IaIo_sig_exp, left_on='HOG', right_on='FamilyID', how='inner')
IaIo_sig_con = pd.merge(HOG, IaIo_sig_con, left_on='HOG', right_on='FamilyID', how='inner')
```

```python
# The evoloved gene families intersecton between I. io and N. aviator
set(IaIo_sig_exp.HomSap) & set(NycAvi_sig_con.HomSap)
```

```python
# save the results 
NycAvi_sig_exp.to_csv('NycAvi_sig_exp.tsv', header=True, index=False)
NycAvi_sig_con.to_csv('NycAvi_sig_con.tsv', header=True, index=False)

IaIo_sig_exp.to_csv('IaIo_sig_exp.tsv', header=True, index=False)
IaIo_sig_con.to_csv('IaIo_sig_con.tsv', header=True, index=False)
```

### potential gene loss events

```python
# count the number of member in each species in each gene families
count = pd.read_csv('./HOG.tsv', header=0, sep='\t')
count.drop(columns=['OG', 'Gene Tree Parent Clade'], inplace=True)
count.fillna(0, inplace=True)
for i in count.columns[1:]:
    count[i] = count[i].apply(lambda x: 0 if x == 0 else x.count('|'))
count
```

```python
# If the number of members in a gene family of a species is 0, it is considered a potential gene loss event
# Identify potential gene loss events in two avivorous bat species
gene_loss = count[(count.apply(lambda x: len(x[x==0]), axis = 1) == 2) & (count.IaIo == 0) & (count.NycAvi == 0)]
gene_loss = pd.merge(HOG, gene_loss, on='HOG', suffixes=['_anno', ''], )
gene_loss
```
