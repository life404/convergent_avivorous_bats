```r
APOA1 = plotRers(mamRERw, 'OG0005604', phenv = phenv, plot=0)
```

```r
APOA1$data %>% ggplot(aes(y = species, x = rer, color = mole)) + 
	geom_vline(xintercept = 0, linewidth = 1.1, color = 'grey80', linetype = 'dashed') +
	geom_point(size = 3) + 
	geom_text(aes(label = species, x = rer - 1)) + 
	theme(
		panel.background = element_blank(), 
		axis.line.x = element_line(color='black', linewidth = 1.1), 
		axis.line.y = element_blank(), axis.text.y = element_blank(), 
		axis.text.x = element_text(color = 'black', size = rel(1.7)), 
		panel.grid = element_blank(), 
		axis.ticks.y = element_blank(),
		axis.title = element_text(color = 'black', size = rel(1.9))
	) + 
	ylab('Branches') + xlab('Relative of Evolution Rate') + 
	scale_color_manual(values = RColorBrewer::brewer.pal(3, 'Set1')[c(2,1)]) 
```

```r
CEPT1 = plotRers(mamRERw, 'OG0012149', phenv = phenv, plot=0)
pdf('CEPT.RER.pdf', width = 6, height = 6)
CEPT1$data %>% ggplot(aes(y = species, x = rer, color = mole)) + 
	geom_vline(xintercept = 0, linewidth = 1.1, color = 'grey80', linetype = 'dashed') +
	geom_point(aes(size = mole)) + 
	geom_text(aes(label = species, x = rer - 0.3)) + 
	theme(
		panel.background = element_blank(), 
		axis.line.x = element_line(color='black', linewidth = 1.1), 
		axis.line.y = element_blank(), axis.text.y = element_blank(), 
		axis.text.x = element_text(color = 'black', size = rel(1.7)), 
		panel.grid = element_blank(), 
		axis.ticks.y = element_blank(),
		axis.title = element_text(color = 'black', size = rel(1.9)),
		legend.position = 'none'
	) + 
	ylab('Branches') + xlab('Relative of Evolution Rate') + 
	scale_color_manual(values = RColorBrewer::brewer.pal(3, 'Set1')[c(2,1)])  + 
	scale_size_manual(values = c(3,5))
dev.off()
```
